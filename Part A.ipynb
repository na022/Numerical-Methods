```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def funf(x, ld, mu):
   f= np.sqrt(ld / (2 * np.pi * x ** 3)) * np.exp(-ld / (2 * mu ** 2 * x) * (x - mu) ** 2)
   return f

# Plot the PDFs
def plotfunf ():
    plt.figure(figsize=(10, 6))
    parameter = [(0.5, 1), (1, 3), (5, 1), (10, 3)]
    color = ['b', 'g', 'r', 'c']
    label = [f'(={ld}, ={mu})' for ld, mu in parameter]
    x = np.linspace(0.01, 5, 1000)

    for (ld, mu), color, label in zip(parameter, color, label):
        f = funf(x, ld, mu)
        plt.plot(x, f, color, label=label)
        plt.xlabel('x', fontsize=16)
        plt.title('PDF on [0, 5]',fontsize=18,fontweight= 'bold' )
        plt.xticks(fontsize=14)
        plt.grid(True)
        plt.legend()


    plt.show()

plotfunf()
```
